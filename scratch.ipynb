```python
from llm import LLM
```

```python
client = LLM(provider="mistral", model_name="mistral-large-latest")
client.query_llm("Who are you?")
```

```python
client.query_structured_llm("Generate a JSON that represents a person, with name and age. Return the JSON object only, no other text.", response_format={"type": "json_object"})
```

```python
client = LLM(provider="cohere", model_name="command-a-03-2025")
client.query_llm("Who are you?")
```

```python
response_format = {
        "type": "json_object",
    }

client.query_structured_llm("Generate a JSON that represents a person, with name and age", response_format=response_format)
```
